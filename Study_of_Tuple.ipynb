```python
#Python tuple
#create tuple and access tuple items
thistuple = ("orange", "grapes", "kiwi")
print(thistuple[1])
```

```python
thistuple = ("orange", "grapes", "kiwi")
thistuple[1]= "blackcurrent"
print(thistuple)
```

```python
#Append items
thistuple = ("orange", "grapes", "kiwi")
thistuple.append("orange")
print(thistuple)
```

```python
#Add tuple to a tuple
thistuple = ("orange", "grapes", "kiwi")
y = ("apple",)
thistuple += y
print(thistuple)
```

```python
#Slicing
numbers = (3, 7, 1, 9, 4, 2, 5)
print(numbers[1::2])
print(numbers[1:3])
print(numbers[1:3:1])
print(numbers[1:5:2])
print(numbers[1:4:3])
print(numbers[6:4:-1])
print(numbers[-2:-4:-2])
```

```python
#Result analysis
result = ("Maths", 85, "A")

# Unpack the tuple
subject, marks, grade = result

#Display the reuslt details
print("Subject:", subject)
print("Marks:", marks)
print("Grade:", grade)

#Check for distinction
if marks >=75:
  print("Distinction")
```

```python
#Attendance
attendance = ("P", "P", "P", "A", "P", "A", "A", "P")
print("Total number of present days:", attendance.count("P"))
print("Total number of absent days:", attendance.count("A"))

if attendance.count("A") >=1:
  print("The employee was absent at least once")
```
